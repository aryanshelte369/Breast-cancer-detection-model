```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

#Define Paths and Load Data
jpeg_root = r"C:\Users\aryan\OneDrive\Desktop\DATASET\jpeg"
csv_path = r"C:\Users\aryan\OneDrive\Desktop\DATASET\csv\mass_case_description_train_set.csv"


if not os.path.exists(csv_path):
    print(f"Error: The CSV file was not found at {csv_path}")
else:
    df = pd.read_csv(csv_path)

    #Function to Get the Parent UID Folder Path
    def get_uid_parent_folder_path(row, root_dir):

        full_image_relative_path = row['image file path']
        roi_mask_relative_path = row['ROI mask file path']
        
        # Get the parent directory of the file path, then get its name (the UID)
        full_uid_folder = os.path.basename(os.path.dirname(full_image_relative_path))
        roi_uid_folder = os.path.basename(os.path.dirname(roi_mask_relative_path))
        
        # Join the root directory with ONLY that UID folder
        full_path = os.path.join(root_dir, full_uid_folder)
        roi_path = os.path.join(root_dir, roi_uid_folder)
        
        return full_path, roi_path

    #Function to Display Images with Specific Title
    def display_images_in_folder(folder_path):

        if not os.path.isdir(folder_path):
            print(f"Error: Folder not found: {folder_path}")
            return
        
        # Get the UID from the folder path for the title
        uid = os.path.basename(folder_path)
        
        # Find all valid image files in the directory
        image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.jpg', '.jpeg'))]
        
        if not image_files:
            print("No JPG or JPEG files found in this folder.")
            return

        # Display each image
        for filename in image_files:
            file_path = os.path.join(folder_path, filename)
            plt.figure(figsize=(6, 8)) # Adjusted for a typical mammogram aspect ratio
            
            try:
                img = Image.open(file_path)
                plt.imshow(img, cmap='gray')
            except Exception as e:
                print(f"Could not read image file {filename}. Error: {e}")
                continue

            # Set the title to the desired format
            plt.title(f"{uid} - {filename}", fontsize=14)
            plt.axis('off')
            plt.show()

    #Main Execution Block
    # Get the first row as a sample
    sample_row = df.iloc[0]

    # Get the desired folder paths
    full_mammogram_folder, roi_mask_folder = get_uid_parent_folder_path(sample_row, jpeg_root)

    # Print the paths and file lists exactly as in the screenshot
    print(f"Full mammogram JPEG folder: {full_mammogram_folder.replace(os.sep, '/')}")
    print(f"ROI mask JPEG folder: {roi_mask_folder.replace(os.sep, '/')}")
    print("\nFull mammogram images:", os.listdir(full_mammogram_folder) if os.path.isdir(full_mammogram_folder) else "Folder not found")
    print("All ROI mask images:", os.listdir(roi_mask_folder) if os.path.isdir(roi_mask_folder) else "Folder not found")

    # Print the header for the image display
    print("\nShowing Full Mammogram Images:")
    display_images_in_folder(full_mammogram_folder)

    print("\nShowing All ROI Mask Images:")
    display_images_in_folder(roi_mask_folder)


# Data Visualization


```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns
import numpy as np
from IPython.display import display

#path to dataset
BASE_DIR = r"C:\Users\aryan\OneDrive\Desktop\DATASET"
csv_dir = os.path.join(BASE_DIR, 'csv')
if not os.path.exists(csv_dir):
    raise FileNotFoundError(f"CSV file {csv_dir} does not exist")
calc_train_path = os.path.join(csv_dir, 'calc_case_description_train_set.csv')
mass_train_path = os.path.join(csv_dir, 'mass_case_description_train_set.csv')
calc_test_path = os.path.join(csv_dir, 'calc_case_description_test_set.csv')
mass_test_path = os.path.join(csv_dir, 'mass_case_description_test_set.csv')
dicom_info_path = os.path.join(csv_dir, 'dicom_info.csv')
meta_path = os.path.join(csv_dir, 'meta.csv')

# Loading DataFrames
calc_train_df = pd.read_csv(calc_train_path)
mass_train_df = pd.read_csv(mass_train_path)
calc_test_df = pd.read_csv(calc_test_path)
mass_test_df = pd.read_csv(mass_test_path)
dicom_info_df = pd.read_csv(dicom_info_path)
meta_df = pd.read_csv(meta_path)

#Dataset shapes
print(f"Calc train set: {calc_train_df.shape}")
print(f"Mass train set: {mass_train_df.shape}")
print(f"Calc test set: {calc_test_df.shape}")
print(f"Mass test set: {mass_test_df.shape}")
print(f"DICOM info: {dicom_info_df.shape}")
print(f"META data: {meta_df.shape}")

#Info on columns
print("\nCalc Train Set Columns Info:", calc_train_df.columns.tolist())
print("\nMass Train Set Columns Info:", mass_train_df.columns.tolist())
print("\nCalc Test Set Columns Info:", calc_test_df.columns.tolist())
print("\nMass Test Set Columns Info:", mass_test_df.columns.tolist())

#Assuming BASE_DIR is already defined pointing to dataset root folder

def normalize_pathology(label):
    if pd.isna(label):
        print("Warning: Missing pathology label found.")
        return None
    if label == 'MALIGNANT':
        return 'malignant'
    return 'benign'

def get_jpeg_image_path(row):
    try:
        parts = row['image file path'].split('/')
        folder_name = parts[-2]
        folder_path =os.path.join(BASE_DIR, 'jpeg', folder_name)
        print(f"Looking for image folder = {folder_path}")
        if not os.path.exists(folder_path):
            print(f"Image folder does not exists: {folder_path}")
            return None
        
        jpg_files = [f for f in os.listdir(folder_path) if f.endswith('.jpg')]
        if len(jpg_files) == 0:
            print(f"No JPEG files found in image folder: {folder_path}")
            return None
        image_path = os.path.join(folder_path, jpg_files[0])
        print(f"Found image file: {image_path}")
        return image_path
    except Exception as e:
        print(f"Error in get_jpeg_image_path: {e}")
        return None
    
def is_mask_image(img_path, threshold_unique_values=20):
    try:
        img = Image.open(img_path).convert('L')
        arr = np.array(img)
        unique_vals = np.unique(arr)
        print(f"Checking mask image: {img_path}  unique values count: {len(unique_vals)}")
        return len(unique_vals) <= threshold_unique_values
    except Exception as e:
        print(f"Error in is_mask_image: {e}")
        return False
    
def get_jpeg_mask_path(row):
    try:
        part = row['ROI mask file path'].split('/')
        folder_name = part[-2]
        folder_path = os.path.join(BASE_DIR, 'jpeg' , folder_name)
        print(f'Looking for mask folder = {folder_path}')
        if not os.path.exists(folder_path):
            print(f"Mask folder does not exists: {folder_path}")
            return None
        jpeg_files = [f for f in os.listdir(folder_path) if f.endswith('jpg')]
        for jpeg_files in sorted(jpeg_files):
            full_path = os.path.join(folder_path, jpeg_files)
            if is_mask_image(full_path):
                print(f'Found mask file: {full_path}')
                return full_path
        fall_back_path =os.path.join(folder_path, jpeg_files[0]) if jpeg_files else None
        print('No mask found by unique values heuristic, fallback to: {fall_back_path}')
        return fall_back_path
    except Exception as e:
        print(f'Error in get_jpeg_mask_path: {e}')
        return None
    
def preprocessing_df(df):
    df = df.copy()
    df['label'] = df['pathology'].apply(normalize_pathology)
    df['image_path'] = df.apply(get_jpeg_image_path, axis = 1)
    df['mask_path'] = df.apply(get_jpeg_mask_path, axis = 1)
    before_drop = len(df)
    df = df.dropna(subset = ['label', 'image_path', 'mask_path'])
    after_drop = len(df)
    print(f"Dropped {before_drop - after_drop} rows due to missing labels or paths.")
    df = df[df['image_path'].apply(os.path.exists)]
    df = df[df['mask_path'].apply(os.path.exists)]
    print(f"Rows after verifying image an dmask paths exists: {len(df)}")
    return df

# Combining calcification and mass datasets for train and test with preprossing
print("\nPreproccessing calc_train_df...")
calc_train_preprocessed = preprocessing_df(calc_train_df)

print("\nPreproccessing mass_train_df...")
mass_train_preprocessed = preprocessing_df(mass_train_df)
train_df = pd.concat([calc_train_preprocessed, mass_train_preprocessed], ignore_index=True)

print("\nPreproccessing calc_test_df...")
calc_test_preprocessed = preprocessing_df(calc_test_df)

print("\nPreproccessing mass_test_df...")
mass_test_preprocessed = preprocessing_df(mass_test_df)
test_df = pd.concat([calc_test_preprocessed, mass_test_preprocessed], ignore_index=True)

# check label distribution after normallization
print("\nTrain set label distribution after normallization:")
print(train_df['label'].value_counts())

sns.countplot(x = 'label', data = train_df)
plt.title('Train set Label Distribution(Binary):')
plt.xlabel('Label')
plt.ylabel('Count')
plt.xticks(rotation = 45)
plt.show()

#Sample images and mask visualization function

def show_samples_with_masks(df, label, n = 4, random_state = 42):
    samples = df[df['label'] == label].sample(n, random_state = random_state)
    plt.figure(figsize = (12,6))

    for i, (_,row) in enumerate(samples.iterrows()):
        try:
            print(f"Showing samples {i+1} for label '{label}':")
            print(f"Image path: {row['image_path']}")
            print(f"Mask path: {row['mask_path']}")

            img =  Image.open(row['image_path'])
            plt.subplot(2, n, i + 1)
            plt.imshow(img, cmap = 'gray')
            plt.title(f"{label}\nImages")
            plt.axis('off')

            mask = Image.open(row['mask_path'])
            plt.subplot(2, n, i + 1 + n)
            plt.imshow(mask, cmap = 'gray')
            plt.title("Mask")
            plt.axis('off')

        except Exception as e:
            print(f"Error displaying sample {i+1} for label '{label}': {e}")

    plt.tight_layout()
    plt.show()

# Display sample images and masks for each label
for lbl in train_df['label'].unique():
    print(f"\nSample images and masks for class: {lbl}")
    show_samples_with_masks(train_df, lbl)

# Show sample paths after preprocessing
print("\nSample paths after preprocessing:")
display(train_df[['image file path', 'image_path', 'mask_path', 'label']].head())

#Define a path for the processed data
processed_data_path = 'processed_train_data.csv'

#Save the dataframe to a CSV file
train_df.to_csv(processed_data_path, index=False)

print(f"DataFrame saved to {processed_data_path}")
```
